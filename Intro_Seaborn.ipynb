## Introduction to Seaborn

Seaborn is an open-source Python library built on top of matplotlib. It is used for data visualization and exploratory data analysis. Seaborn works easily with dataframes and the Pandas library. The graphs created can also be customized easily.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python

import numpy as np

xpoints = np.array([1, 8])
ypoints = np.array([3, 10])
plt.xlabel("xpoints")
plt.ylabel('ypoints')
plt.title('X vs Y')
plt.grid(axis='y')
plt.plot(xpoints, ypoints)
plt.show()
```

```python
## Seaborn has inbuilt datasets which can be used for practice purpose.
sns.get_dataset_names()
```

```python
data=sns.load_dataset('titanic')
data.head()
```

```python
data.dtypes
```

While dealing with data we have categorical as well as numerical data and while trying to plot them there are various types of graph for each one and Seaborn provides simple syntax to plot as well as customise the graphs. 
Let's look into different types of plots -

## Categrical Plot
### Countplot 

```python
sns.countplot(x='sex',data=data)
```

```python
## customize the graph
sns.countplot(x='sex',data=data,hue='survived',palette="Pastel1")
```

```python
sns.countplot(x='sex',data=data,hue='survived',palette="Set1")
```

## Distribution Plot
### KDE(Kernel Density Estimate)

```python
sns.kdeplot(x = 'age' , data = data , color = 'red')
```

### Displot-plots distribution of continuous data.

```python
sns.displot(x='age',data=data,kde=True,bins=5)
```

## Relational Plot
### Scatterplot - helps understand relation between data.


```python
df=sns.load_dataset('iris')
df.head()
df.dtypes
```

```python
sns.scatterplot(x='sepal_length',
               y='petal_length',
               data=df,
               hue='species')
```

```python
sns.set_style('darkgrid')
sns.scatterplot(x='sepal_length',
               y='petal_length',
               data=df,
               hue='species',
               )
sns.despine()
```

## Jointplot

```python
sns.jointplot(x='sepal_length' , y ='petal_length',
data = df , kind = 'reg')
```

```python
sns.jointplot(x='sepal_length' , y ='petal_length',
data = df , kind = 'hex')
```

### Pairplot- Seaborn lets us plot multiple scatter plots. Its a good option when you want to get a quick overview of your data.

```python
sns.pairplot(df)
```

### Heatmaps - Used to visualise confusion matrix and correlations

```python
corr=df.corr()
sns.heatmap(corr)
```

## FacetGrid

```python
fg1=sns.FacetGrid(df,col='species')
fg1.map(plt.hist,'sepal_length')
```

```python
fg2=sns.PairGrid(df)
fg2.map(plt.scatter)
```
